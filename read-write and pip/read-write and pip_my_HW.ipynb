```python
# Задание 1. Переведите содержимое файла purchase_log.txt в словарь purchases вида:
# {‘1840e0b9d4’: ‘Продукты’, …}
# содержимое purchase_log.txt помещается в оперативную память компьютера
```

```python
# Вариант 1
import json

file_pl_name = 'purchase_log.txt'
file_out_name = 'purchases.txt'
purchases = {}
i = 0

with open(file_pl_name, 'r', encoding='utf-8') as file_pl:
    for line in file_pl:
        line = line.strip()
        dict_pl = json.loads(line)
        new_key = dict_pl['user_id']
        new_value = dict_pl['category']
        purchases[new_key] = new_value
        i += 1

# на экран не выводит, т.к. не хватает памяти, ошибка 'IOPub data rate exceeded'
# поэтому вместо 'print' всего словаря далее вывожу первые 19 (20 с "шапкой") пар "ключ: значение"
# и также доаполнительно делаю выгрузку всего словаря в файл
# print(purchases.items())
        
# вывод на экран с граничением количества данных для контроля
for n, (k, v) in enumerate(purchases.items()):
    while n < 20:
        print(k, v)
        break
        
# вывод в файл 'purchases.txt' для контроля
with open(file_out_name, 'w') as file_out:
    txt_out = json.dumps(purchases)
    file_out.write(txt_out)
    
print(i) # 100000
```

```python
# Вариант 2
import json

file_pl_name = 'purchase_log.txt'
file_out_name = 'purchases.txt'
purchases = {}
temp_ = set()
i = 0

with open(file_pl_name, 'r', encoding='utf-8') as file_pl:
    for line in file_pl:
        line = line.strip()
        dict_pl = json.loads(line)
        new_key = dict_pl['user_id']
        new_value = dict_pl['category']
        
#         проверяем на дубликаты и не обрабатываем их
        if not new_key in temp_:
            purchases[new_key] = new_value
            temp_.add(new_key)
            i += 1

# вывод в файл 'purchases.txt' для контроля
with open(file_out_name, 'w') as file_out:
    txt_out = json.dumps(purchases)
    file_out.write(txt_out)

print(i) # 99518
```

```python
# Вариант 1.
# Доработано после первичной проверки

import json

file_pl_name = 'purchase_log.txt' # 11920
file_vl_name = 'visit_log.csv' # 59117
file_out_name = 'funnel_v1.csv'
file_p = 'purchases.txt'
i = 0

with open(file_p, 'r', encoding='utf-8') as p:
     purchases = json.load(p)

with open(file_vl_name, 'r', encoding='utf-8') as file_vl:
    with open(file_out_name, 'w') as file_out:
        for n, line_vl in enumerate(file_vl): # цикл по файлу 'visit_log.txt'
            user_id = line_vl.strip().split(',')[0]
            if user_id in purchases:
                i += 1
                file_out.write(line_vl.strip() + ',' + purchases[user_id] + '\n')
print(i) # 13799, но там 265 (169) задвоений
```
