```python
#  Вводные

documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
# перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

```python
'''============================
          Функции
==============================='''

def menu_choice(command_list):
  print('Достуные команды:')
  for (k, v) in commands_list.items():
    print(k, v[0], sep=' - ')

def owner_name(doc_numb):
  # номер документа --> владелец документа 
  owner_name = [doc_line['name'] for doc_line in documents if doc_line['number'] == user_doc_number]
  return "".join(owner_name)

def dir_number(doc_numb):
  # номер документа --> полка хранения
  dir_number = [direc for direc, numb in directories.items() if doc_numb in numb]
  return "".join(dir_number)

def full_info():
  # полный перечень существующих документов
  return ('\n'.join(['№: ' + doc_line['number'] + ', тип: ' + doc_line['type']  + ', владелец: ' +
          doc_line['name'] + ', полка хранения: ' + dir_number(doc_line['number']) 
          for doc_line in documents]))

def add_dir(ads):
  if ads in directories.keys():
    print('Такая полка уже существует.', end=' ')
  else:
    directories[ads] = None
    print('Полка добавлена.', end=' ')
  shelfs()

def shelfs():
  print('Текущий перечень полок:', ", ".join(directories.keys()))

# удаление полки
def del_dir(ds): 
  if not ds in directories.keys():
    print('Такой полки не существует.', end=' ')
  elif len(directories[ds]) > 0:
    print('На полке есть документы, удалите их перед удалением полки.', end=' ')
  elif len(directories[ds]) == 0:
    del directories[ds]
    print('Полка удалена.', end=' ')
  # print('Текущий перечень полок:', ", ".join(directories.keys()))
  shelfs()

def no_doc():
  # print('Документ не найден в базе', end=' ')
  print('Документ не найден в базе')

def no_shelf():
  print('Такой полки не существует.', end=' ')

# есть ли документ
def ifa_doc(doc_numb):
  print(doc_numb in directories.values())
  return doc_numb in directories.values()

# есть ли полка
def ifa_shelf(doc_shelf):
  return doc_shelf in directories.keys()

# добавление документа
def add_doc():
  add_doc_number = input('Введите номер документа: ')
  add_doc_type = input('Введите тип документа: ')
  add_doc_owner = input('Введите владельца документа: ')
  add_doc_shelf = input('Введите полку для хранения: ')
  if not ifa_shelf(add_doc_shelf):
    no_shelf()
    print('Добавыьте полку командой "as"')
  else:
    documents.append({'type': add_doc_type, 'number': add_doc_number, 'name' : add_doc_owner})
    directories[add_doc_shelf].append(add_doc_number)
    print('Документ добавлен.', end=' ')
  print('Текущий список документов: \n' + full_info())

# удаление документа
def del_doc(del_doc_number):
  if dir_number(del_doc_number) == '':
    no_doc()
  else:
    # убираем документ из списка
    doc_seq = [n for n, x in enumerate(documents) if documents[n]['number'] == '10006']
    doc_seq = int(''.join(str(a) for a in doc_seq))
    documents.pop(doc_seq)
    # убираем с полки
    shlf = (dir_number(del_doc_number))
    directories[shlf].remove(del_doc_number)
    print('Документ удален.')
  print('Текущий список документов: \n' + full_info())



'''============================
          Основной код
==============================='''

# the start is right here

# перечень доступных команд
commands_list = {
  '  p':['поиск владельца документа по номеру', 'owner_name', 'user_doc_number'],
  '  s':['поиск полки по номеру документа'],
  '  l':['вывод полной информации о документах в БД'],
  '  as':['добавление новой полки'],
  '  ds':['удаление существующей полки'],
  '  ad':['добавление нового документа'],
  '  d':['удаление существующего документа'],
  '  m':['перемещение существующего документа на другую полку']
}

menu_choice(commands_list)

# commands_list = ['p', 's', 'l', 'as', 'ds', 'ad', 'd', 'm']

user_command = input('Введите команду: ')

# поиск владельца документа по номеру
if user_command == 'p':
  user_doc_number = input('Введите номер документа: ')
  o_name = owner_name(user_doc_number)
  if o_name == '':
    no_doc()
  else:
    print(f'Владелец документа: {o_name}')
# поиск полки по номеру
elif user_command == 's':
  user_doc_number = input('Введите номер документа: ')
  d_num = dir_number(user_doc_number)
  if d_num == '':
    no_doc()
  else:
    print(f'Документ хранится на полке: {d_num}')
# вывод перечня всех документов
elif user_command == 'l': # буква 'L.lower()'
  # print("\n".join(full_info()))
  print(full_info())
# добавление новой полки
elif user_command == 'as':
  add_dir(input('Введите номер полки: '))
# удаление полки
elif user_command == 'ds':
  del_dir(input('Введите номер полки: '))
# добавление документа
elif user_command == 'ad':
  add_doc()
# удалние документа
elif user_command == 'd':
  del_doc(input('Введите номер документа: '))
#  перенос документа
elif user_command == 'm':
    print('Функционал будет доступен позднее')
#   move_doc(input('Введите номер полки: '))
elif user_command not in commands_list:
  print('Команда не распознана, выход')
  quit()
```
